# **Medidas de Tendência Central**

São conhecidas por estudar o que está acontecendo com os dados que estão no entorno de um ponto e precisamos de uma medida que indique o ponto de equilíbrio desses dados

Dados populacionais = paramestros

Dados amostrais = estimadores ou estatísticas

Principais medidas
Média

Moda

Mediana 

Percentis


**Média Aritmética** 

Soma de todos os elementos dividida pelo número total de observações

CARACTERISTICAS

facilmente encontrada

bastante sensível as alteraçõces dos dados

a soma da diferença de cada valor obserbador em relaçaõ a media é zero, ou seja, a soma dos desvios é zero

Pode ser afetadas por valores extremos (outliars)

**Moda**

Representa a maior frequencia da variável entre os valores observados.

Ex: A = {1,2,5,2,1,2,1,5,7,1,2,1,1,1}
Mod=1

**Mediana**

É o valor que o ocup a posição central dos dados ordenados

Ex: A={12,22,55,2,1} valores - impar

Aord={1,2,12,22,55}

Med=12

Ex: B={1,2,5,2,1,2,1,5,7,1,2,1,1,1} valores - par

Bord={1,1,1,1,1,1,1,2,2,2,2,5,5,7}

Med=(1+2)/2 = 1.5


Medidas em Conjunto = Dispersão

* Assimetria a esquerda (media, mediana, moda)

* Simétrica (media, mediana, moda)

* Assimetria a direita (moda, mediana, media)


**Medidas Separaratrizes**

Medidas que dividem o conjunto de dados em partes iguais, podem ser:

* Quartil: divide o conjunto de dados em 4 partes iguais

p=0,25(n+1)

p=0,50(n+1)

p=0,75(n+1)

* Decil:divide o conjunto de dados em 10 partes iguais
* Percentildivide o conjunto de dados em 100 partes iguais

p=0,05(n+1)

p=0,10(n+1)

p=0,25(n+1)

p=0,50(n+1)

p=0,75(n+1)

p=0,90(n+1)

p=0,95(n+1)

Os dados devem estar ordenados.

# **Medidas de Dispersão**

Dados apresentam semelhanças e variabilidades
Permitem identificar ate que ponto os resultados se concentram ou não ao redor da tendência central de um conjunto de observações
As medidas de dispersão auxiliam as medidas de tendencia central a descrever melhor os dados
Indicam o quanto os dados estão ou não próximos uns dos outros
Não há sentido em calcular a média de um conjunto de dados onde não há variação dos seus elementos
Precisamos de uma MTC e MD para descrever adequadamente dados.

**Amplitude Total**

É a diferença entre o maior e o menor valor observado
Não leva em considerçaão valores intermediários
Não temos informações de como os dados estão distribuidos

Ex: w={12,23,65,11,3,9,32}
Amp = 65-3 = 62


**Amplitude Interquartílica**

Diferença entre o primeiro e o terceiro quartil
Medida mais estável do que a amplitude total
Abrange 50% dos dados
É util para detectar valores discrepantes - outliars

dq = Q3 - Q1

Amplitude semi-interquartilica
dqm = (Q3 - Q1)/2


**Desvio Médio**

A diferença entre cada valor observado e a média

Xi - u

XI - ~X

A soma dos desvios é igual a Zero (preciso tirar o módulo)
Devemos desconsiderar os sinais utilizando o módulo e a média dessas diferenças é denominada desvio médio

**Variância e Desvio Padrão**

Algumas propriedades matemáticas para o desvio médio possuem desvantagens
Por isso utilizamos a variância
Retorna a dimensão original dos dados

**Coeficiente de Variação**

Indica variabilidade da medida em relação a media
Vantagem: permite comparar a dispersão de diferentes distribuições com diferentes médias e desvios padrões

**Medidas de Curtose**

Quantifica a concentração ou dispersão de valores de um conjunto de dados em relação às medidas de tendência central
Mede o grau de achatamento de uma distribuição
Complementar as informações de dispersão

Classificação de grau de achatamento

Leptocurtica: dados muito concentrados em torno do centro (K < 0.263)
Mesocúrtica: dados LEVEMENTE concentrados em torno do centro (K = 0.263)
Platicúrtica dados pouco concentrados em torno do centro (K > 0.263)
